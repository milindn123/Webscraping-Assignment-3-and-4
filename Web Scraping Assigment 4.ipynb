9. Scrape the details most watched tv series of all time from imdb.com.
Url = https://www.imdb.com/list/ls095964455/
You have to find the following details:
A) Name
B) Year span
C) Genre
D) Run time
E) Ratings
F) Votes

```python
#sending request to web page
url= "https://www.imdb.com/list/ls095964455/"
page=requests.get(url)
```

```python
soup=BeautifulSoup(page.content)
```

```python
name=soup.find_all('h3', class_="lister-item-header")
tvshow_name=[]
for i in name:
    for j in i.find_all("a"):
        tvshow_name.append(j.text)
    
tvshow_name    
```

```python
year=soup.find_all('span', class_="lister-item-year text-muted unbold")
span_year=[]
for i in year:
    span_year.append(i.text)
    
    
    
    
span_year 
```

```python
genre=soup.find_all('span', class_="genre")
tvshow_genre=[]
for i in genre:
      tvshow_genre.append(i.text.replace("\n",''))
  
    
tvshow_genre    
```

```python
time=soup.find_all('span', class_="runtime")
run_time=[]
for i in time:
      run_time.append(i.text)
  
    
run_time 
```

```python
rating=soup.find_all('span', class_="ipl-rating-star__rating")
show_rating=[]
for i in range(0,len(rating),23):
    show_rating.append(rating[i])
    
show_rating        
        
            
        
    

    
```

```python
vote=soup.find_all('p',class_="text-muted text-small")
show_vote=[]
for i in vote:
    show_vote.append(i.text.replace('\n',''))
        
  
    
show_vote
```

```python
votes=[]
for i in range(2,len(show_vote),3):
    votes.append(show_vote[i])
votes    
```

```python
print(len(votes),len(show_rating),len(run_time),len(tvshow_genre),len(span_year),len(tvshow_name))
```

```python
mostwatched_tvseries=pd.DataFrame({})
mostwatched_tvseries['Show Name']=tvshow_name
mostwatched_tvseries['Year Span']=span_year
mostwatched_tvseries['Genre']=tvshow_genre
mostwatched_tvseries['Run Time']=run_time
mostwatched_tvseries['Ratings']=show_rating
mostwatched_tvseries['Show Vote']=votes
mostwatched_tvseries
```

```python
genr=[]
for i in range(5, len(titles), 6):
    genr.append(titles[i])
genr
```

```python
print(len(genr),len(publ),len(v_s),len(a_n),len(b_t))
```

7. Scrape the details of Data science recruiters from naukri.com.
Url = https://www.naukri.com/

A) Name
B) Designation
C) Company
D) Skills they hire for
E) Location

```python
url2='https://www.naukri.com/data-science-recruiters'
page2=requests.get(url2)
soup2=BeautifulSoup(page2.content)
```

```python
name=soup2.find_all('span', class_="fl ellipsis")
recruiter_name=[]
for i in name:
    recruiter_name.append(i.text)
recruiter_name
```

```python
designation=soup2.find_all('span', class_="ellipsis clr")
recruiter_designation=[]
for i in designation:
    recruiter_designation.append(i.text)
recruiter_designation
```

```python
company=soup2.find_all('a', class_="ellipsis")
recruiter_company=[]
for i in company:
    recruiter_company.append(i.text)
recruiter_company
rec_comp=[]
for i in range(1, len(recruiter_company), 2):
    rec_comp.append(recruiter_company[i])
rec_comp    
```

```python
skills=soup2.find_all('div', class_="hireSec highlightable")
recruiter_skills=[]
for i in skills:
    recruiter_skills.append(i.text)
recruiter_skills
```

```python
location=soup2.find_all('small', class_="ellipsis")
recruiter_location=[]
for i in location:
    recruiter_location.append(i.text)
recruiter_location
```

```python
recruiter_location.append("Delhi")
```

```python
print(len(recruiter_name),len(recruiter_designation),len(rec_comp),len(recruiter_skills),len(recruiter_location))  
```

```python
data_sci_recruiter=pd.DataFrame({})
data_sci_recruiter['Name']=recruiter_name
data_sci_recruiter['Designation']=recruiter_designation
data_sci_recruiter['Company']=rec_comp
data_sci_recruiter['Skills']=recruiter_skills
data_sci_recruiter['Location']=recruiter_location
data_sci_recruiter
```

```python
lang_used=soup4.find_all('span', class_="d-inline-block ml-0 mr-3") 
l_n=[]
for i in lang_used:
    l_n.append(i.text.replace('\n',''))
l_n
```
